```python
import requests
import zipfile
import os
import pandas as pd
root = 'https://shoonya.finvasia.com/'
#For FnO token download
masters = ['NFO_symbols.txt.zip', 'NSE_symbols.txt.zip'] #'CDS_symbols.txt.zip', 'MCX_symbols.txt.zip', 'BSE_symbols.txt.zip' 
#For index data
#masters = ['NSE_symbols.txt.zip', 'NFO_symbols.txt.zip', 'CDS_symbols.txt.zip', 'MCX_symbols.txt.zip', 'BSE_symbols.txt.zip'] 


for zip_file in masters:    
    print(f'downloading {zip_file}')
    url = root + zip_file
    r = requests.get(url, allow_redirects=True)
    open(zip_file, 'wb').write(r.content)
    file_to_extract = zip_file.split()
    
    try:
        with zipfile.ZipFile(zip_file) as z:
            #Extacting the file and saving ito the current working directory
            z.extractall("C:\\Users\\RAJU SINGH\\Y's_Strangle_Streg\\Shoonya_Data_Fetch_API")
            print("Extracted: ", zip_file)
    except:
        print("Invalid file")

    os.remove(zip_file)
    print(f'remove: {zip_file}')
    #break
    
########################################################################################################
### FOR Spot Token (NSE)

#Spot Token filtering
import warnings
warnings.filterwarnings('ignore')

all_token_spot = pd.read_csv('NSE_symbols.txt')

spot_token = all_token_spot[all_token_spot['Instrument']=='INDEX']

spot_token['Symbol']=spot_token['Symbol'].replace(['Nifty 50','Nifty Bank','Nifty Fin Services'],['NIFTY','BANKNIFTY','FINNIFTY'])

spot_token['Expiry'] = 'NaN'
spot_token['OptionType'] = 'NaN'
spot_token['StrikePrice'] = 'NaN'

spot_token.rename(columns = {'Unnamed: 7':'Unnamed: 10'},inplace = True)

spot_token = spot_token[['Exchange', 'Token', 'LotSize', 'Symbol', 'TradingSymbol', 'Expiry',
       'Instrument', 'OptionType', 'StrikePrice', 'TickSize', 'Unnamed: 10']]

nifty_spot = spot_token[spot_token['Symbol']=='NIFTY']
bn_spot = spot_token[spot_token['Symbol']=='BANKNIFTY']
fn_spot = spot_token[spot_token['Symbol']=='FINNIFTY']
#mcp_spot = spot_token[spot_token['Symbol']=='MIDCPNIFTY']

###################################################################################################
### FOR Derivative TOKEN (NFO)

# Token filtering

all_token = pd.read_csv('NFO_symbols.txt')

option_token = all_token[all_token['Instrument']=='OPTIDX'].reset_index(drop=True)

#list(option_token[option_token['Symbol']=='MIDCPNIFTY'].index)

#option_token = option_token.drop(list(option_token[option_token['Symbol']=='MIDCPNIFTY'].index)).reset_index(drop=True)

nifty_option_token = option_token[option_token['Symbol']=='NIFTY']
NIFTY_fut_token = all_token[(all_token['Instrument']=='FUTIDX') & (all_token['Symbol']=='NIFTY')].reset_index(drop=True)
#NIFTY_fut_token['Symbol'] = 'NIFTY_fut'

bn_option_token = option_token[option_token['Symbol']=='BANKNIFTY']
BANKNIFTY_fut_token = all_token[(all_token['Instrument']=='FUTIDX') & (all_token['Symbol']=='BANKNIFTY')].reset_index(drop=True)
#BANKNIFTY_fut_token['Symbol'] = 'BANKNIFTY_fut'

fn_option_token = option_token[option_token['Symbol']=='FINNIFTY'].reset_index(drop=True)
FINNIFTY_fut_token = all_token[(all_token['Instrument']=='FUTIDX') & (all_token['Symbol']=='FINNIFTY')].reset_index(drop=True)
#FINNIFTY_fut_token['Symbol']='FINNIFTY_fut'

mcp_option_token = option_token[option_token['Symbol']=='MIDCPNIFTY'].reset_index(drop=True)
MIDCPNIFTY_fut_token = all_token[(all_token['Instrument']=='FUTIDX') & (all_token['Symbol']=='MIDCPNIFTY')].reset_index(drop=True)
#FINNIFTY_fut_token['Symbol']='FINNIFTY_fut'


nifty_token = pd.concat([nifty_option_token,NIFTY_fut_token,nifty_spot]).reset_index(drop=True)
bn_token = pd.concat([bn_option_token,BANKNIFTY_fut_token,bn_spot]).reset_index(drop=True)
fn_token = pd.concat([fn_option_token,FINNIFTY_fut_token,fn_spot]).reset_index(drop=True)
mcp_token = pd.concat([mcp_option_token,MIDCPNIFTY_fut_token]).reset_index(drop=True)


################################################################################################################
#Saving file to current workig directory
from datetime import date
curr_date = str(date.today())
path = os.getcwd()

filename_nfty = "\\nifty_token" + "_" + curr_date + ".txt"
filename_bn = "\\bn_token" + "_" + curr_date + ".txt"
filename_fn = "\\fn_token" + "_" + curr_date + ".txt"
filename_mcp = "\\mcp_token" + "_" + curr_date + ".txt"


nifty_token.to_csv(path + filename_nfty, header =True,index = False )
bn_token.to_csv(path + filename_bn, header =True,index = False )
fn_token.to_csv(path + filename_fn, header =True,index = False )
mcp_token.to_csv(path + filename_mcp, header =True,index = False )
```
